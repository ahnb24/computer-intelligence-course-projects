```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from skimage.io import imread
data = np.vstack([
    np.random.normal(-1 ,0.5 ,size=(100,2)),
    np.random.normal(1 ,0.5 ,size=(100,2)),
    np.random.normal([2 ,1.5] ,size=(100,2)),
    np.random.normal([0 ,-1.5] ,size=(100,2))
])

KMeans1 = KMeans(4).fit(data)
Centroids = KMeans1.cluster_centers_

# plt.scatter(data[:,0],data[:,1],c = KMeans1.labels_ ,cmap = 'rainbow' ,s = 50)
# plt.scatter(Centroids[:,0],Centroids[:,1],c = 'black' ,s = 200 ,marker = 'x')

err = []
for i in range(2,15) :
    KMeans1 = KMeans(i).fit(data)
    err.append(KMeans1.inertia_)
# print(err)
plt.plot(range(2,15),err,"o-")
plt.xlabel("num_cluster")
plt.ylabel("err")
plt.show()
```

```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from skimage.io import imread

image = imread('a.jpg')
image_reshaped = image.reshape((-1,3))
kmeans = KMeans(16).fit(image_reshaped)
labels = kmeans.predict(image_reshaped)
compressed_palette = kmeans.cluster_centers_
reconstructed_image = compressed_palette[labels].reshape(image.shape)
```
