```python
#Experiment2
#Amirhesam_Nakhaee
#Hassan_Farhoud

import numpy as np

def sig(x) :
    y = (2/(1+np.exp(-x))-1)
    return y
def sigdot (x) : 
    y = (1 -  sig(x) ** 2)/2
    return y

class perceptron :
    def __init__(self ,n ) :
        self.w = np.random.random((n,1))*0.01
        

    def forward(self ,x):
        self.x = x
        y = sig(x @ self.w)
        return y

    def grad(self ,label,x,k):
        y_pre = self.forward(x)
        err = label - y_pre
        grad = (np.transpose((sigdot(x) * x))) @ err
        i = 0
        while np.mean(err)  > 0.001 :
            i += 1
            # print(f'mean(err) = {np.mean(err)}')
            self.w += grad*k
            y_pre = self.forward(x)
            err = label - y_pre
        print(f'mean(err) = {np.mean(err)}')
        # print(i)
        j = 0
        for i in y_pre :
            if i > 0.5 :
                y_pre[j] = 1
            else :
                y_pre[j] = 0
            j += 1
        return y_pre
            
ins = perceptron(2)
x = np.array([[0,0],[0,1],[1,0],[1,1]])
label = np.array([[0],[0],[0],[1]])

print(f'\n result : \n{ins.grad(label ,x ,0.1)}')
```

```python
#HW1

import numpy as np

def sig(x) :
    y = (2/(1 + np.exp(-x))) - 1
    return y
def sigdot (x) : 
    y = (1 -  sig(x) ** 2)/2
    return y

class perceptron :
    def __init__(self ,n ) :
        self.w = np.random.random((n+1,1)) + 1
        

    def forward(self ,x):
        bios = np.ones(np.shape(x)[0])
        self.x = np.c_[x,bios]
        y = sig(self.x @ self.w)
        return y

    def grad(self ,label,x,k):
        y_pre = self.forward(x)
        err = y_pre - label
        grad = (np.transpose((sigdot(self.x) * self.x))) @ -err
        i = 0
        while np.mean(err)  > 0.005 :
            # print(f'sum(err) = {sum(err)}')
            i += 1
            self.w += grad*k
            y_pre = self.forward(x)
            err = y_pre - label
        print(f'mean(err) = {np.mean(err)}')
        print(i)
        j = 0
        for i in y_pre :
            if i > 0.5 :
                y_pre[j] = 1
            else :
                y_pre[j] = 0
            j += 1
        return y_pre
            
ins = perceptron(2)
x = np.array([[0,0],[0,1],[1,0],[1,1]])
label = np.array([[0],[0],[0],[1]])

print(f'\n result : \n{ins.grad(label ,x ,0.01)}')
```
